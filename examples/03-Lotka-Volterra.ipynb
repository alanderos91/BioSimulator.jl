# Lotka-Volterra

```julia
using BioSimulator
```

```julia
α = 1.0
β = 0.006 # β + δ = 0.005
δ = 0.001
γ = 0.4

model = Network("Lotka-Volterra")
```

```julia
model <= Species("X", 50)
model <= Species("Y", 50)
```

```julia
model <= Reaction("prey reproduction", α, "X --> X + X")
model <= Reaction("predation", δ, "X + Y --> Y")
model <= Reaction("predator reproduction", β, "X + Y --> Y + Y")
model <= Reaction("predator death", γ, "Y --> 0")
```

```julia
visualize(model)
```

```julia
@time result = simulate(model, SSA, time = 25.0, epochs = 100, trials = 1000)
```

```julia
plot(Trajectory(result, "X", 1))
plot!(Trajectory(result, "Y", 1))
```

```julia
plot(MeanTrajectory(result, "X"), ylim=(0,Inf))
plot!(MeanTrajectory(result, "Y"))
```

```julia
plot(PhaseTrajectory(result, "X", "Y", 1))
```

```julia
plot(Histogram(result, "X", 25.0), xlim=(0,mean(result["X"])), nbins=1000, alpha=0.5)
```

```julia
plot(Histogram(result, "Y", 25.0), alpha=0.5)
```
